## Metrics to measure performance of the model.

**MSE** - Mean Squared Error
$$\text{MSE}(y, \hat{y}) = \frac{1}{n_\text{samples}} \sum_{i=0}^{n_\text{samples} - 1} (y_i - \hat{y}_i)^2.$$
Where:
-  $\hat{y}_i$: predictions of your model
-  $y_i$: true values (targets)

If you want your model to have small errors, `MSE` is a good choice. If you are comparing two or more models with `MSE`, the model with the lowest `MSE` score is better. However:

- Unintuitive
- Heavily effected by outliers

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

**RMSE** - Root Mean Squared Error

$$\text{RMSE}(y, \hat{y}) = \sqrt{\text{MSE}(y, \hat{y})} = \sqrt{\frac{1}{n_\text{samples}} \sum_{i=0}^{n_\text{samples} - 1} (y_i - \hat{y}_i)^2}$$

```python
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(y_test, y_pred)
```

It is the same as `MSE` with a square root. However, error is on the target unit scale.

**MAE** - Mean Absolute Error
$$\text{MAE}(y, \hat{y}) = \frac{1}{n_{\text{samples}}} \sum_{i=0}^{n_{\text{samples}}-1} \left| y_i - \hat{y}_i \right|.$$

- Not penalizes outliers so heavily as `MSE` does
- On the target unit scale

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```

### Relative Metrics

**MAPE** - Mean Absolute Percentage Error

$$\text{MAPE}(y, \hat{y}) = \frac{1}{n_{\text{samples}}} \sum_{i=0}^{n_{\text{samples}}-1} \frac{{}\left| y_i - \hat{y}_i \right|}{\max(\epsilon, \left| y_i \right|)}$$
where:
- $\epsilon$: some small arbitrary but strictly positive number

The idea of this metric is to be sensitive to relative errors. It is for example not changed by a global scaling of the target variable. It is in percentages.

```python
from sklearn.metrics import mean_absolute_percentage_error
y_true = [1, 10, 1e6]
y_preds = [0.9, 15, 1.2e6]
mean_absolute_percentage_error(y_true, y_preds)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test, y_pred)
```

Other metrics you can consider: **SMAPE**, **WAPE**.

# Linear Regression

`Scikit-learn` provides dozens of built-in machine learning algorithms and models, called estimators. Each estimator can be fitted to some data using its *fit* method.

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

The fit method generally accepts 2 inputs:

- The samples matrix (or design matrix) X. The size of `X` is typically `(n_samples, n_features)`, which means that samples are represented as rows and features are represented as columns.

- The target values y which are real numbers for regression tasks, or integers for classification (or any other discrete set of values). For unsupervised learning tasks, `y` does not need to be specified. `y` is usually a 1d array where the `i` th entry corresponds to the target of the i th sample (row) of `X`.

Both `X` and `y` are usually expected to be numpy arrays or equivalent array-like data types, though some estimators work with other formats such as sparse matrices.
Examples of `array-like` datatype:
- a numpy array
- a list of numbers
- a list of length-k lists of numbers for some fixed length k
- a `pandas.DataFrame` with all columns numeric
- a numeric `pandas.Series`

Once the estimator is fitted, it can be used for predicting target values of new data. You don’t need to re-train the estimator:

```python
preds = lr.predict(X_test)
```

```python
print(F"MSE: {mean_squared_error(y_test, preds)}")
print(F"MAE: {mean_absolute_error(y_test, preds)}")
print(F"RMSE: {root_mean_squared_error(y_test, preds)}")
print(F"MAPE: {mean_absolute_percentage_error(y_test, preds)}")
```

## Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
pipe = Pipeline(steps = [
    ('scaler', StandardScaler()),
    ('lr', LinearRegression())
])
pipe.fit(X_train, y_train)
preds = pipe.predict(X_test)
```

```python
print(F"MSE: {mean_squared_error(y_test, preds)}")
print(F"MAE: {mean_absolute_error(y_test, preds)}")
print(F"RMSE: {root_mean_squared_error(y_test, preds)}")
print(F"MAPE: {mean_absolute_percentage_error(y_test, preds)}")
```

### Categorical Variables

If your dataset has columns with several data types, you can use `ColumnTransformer` to preprocess each data type individually.

```python
X_train['OldHouse'] = X_train['HouseAge'] > 26
```

```python
X_train['OldHouse'] = X_train['OldHouse'].astype(int)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
numeric_features = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup',
       'Latitude', 'Longitude']
categorical_features = ['OldHouse']

# Define transformers for each type
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numeric_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])

# Combine preprocessing and model into a pipeline
pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('lr', LinearRegression())
])

pipeline.fit(X_train, y_train)
```
