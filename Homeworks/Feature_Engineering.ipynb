# Feature Engineering

```python
import pandas as pd
import numpy as np

df = pd.read_json('../db/renthop_train.json')
df['created'] = pd.to_datetime(df['created'])
df.head()
```

10 признаков

1. сколько фото — из photos, просто len()

```python
df['num_photos'] = df['photos'].apply(len)
df[['listing_id', 'num_photos']].head(10)
```

2. сколько удобств в списке — из features, len()

```python
df['num_features'] = df['features'].apply(len)
df[['listing_id', 'num_features']].head(10)
```

3. длина описания в символах — из description

```python
df['description_len'] = df['description'].fillna('').apply(len)
df[['listing_id', 'description_len']].head(10)
```

4. квартал (1–4) из created — на паре делали час/месяц, квартал не трогали

```python
df['created_quarter'] = df['created'].dt.quarter
df[['listing_id', 'created', 'created_quarter']].head(10)
```

5. число слов в описании — description.split()

```python
df['description_word_count'] = df['description'].fillna('').apply(lambda s: len(s.split()))
df[['listing_id', 'description_word_count', 'description_len']].head(10)
```

6. есть ли Doorman в features — 0 или 1

```python
df['has_doorman'] = df['features'].apply(lambda x: 1 if 'Doorman' in (x if isinstance(x, list) else []) else 0)
df[['listing_id', 'has_doorman', 'features']].head(10)
```

7. есть ли Elevator в features — 0 или 1

```python
df['has_elevator'] = df['features'].apply(lambda x: 1 if 'Elevator' in (x if isinstance(x, list) else []) else 0)
df[['listing_id', 'has_elevator']].head(10)
```

8. сколько объявлений у этого менеджера — считаю по manager_id

```python
manager_counts = df['manager_id'].value_counts()
df['manager_listings_count'] = df['manager_id'].map(manager_counts)
df[['listing_id', 'manager_id', 'manager_listings_count']].head(10)
```

9. сколько объявлений в этом здании — по building_id

```python
building_counts = df['building_id'].value_counts()
df['listings_per_building'] = df['building_id'].map(building_counts)
df[['listing_id', 'building_id', 'listings_per_building']].head(10)
```

10. студия или нет — bedrooms == 0

```python
df['is_studio'] = (df['bedrooms'] == 0).astype(int)
df[['listing_id', 'bedrooms', 'is_studio']].head(10)
```

все новые столбцы

```python
new_cols = [
    'num_photos', 'num_features', 'description_len', 'created_quarter',
    'description_word_count', 'has_doorman', 'has_elevator',
    'manager_listings_count', 'listings_per_building', 'is_studio'
]
df[new_cols].head(15)
```
