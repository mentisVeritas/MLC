```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
site = "https://www.scrapethissite.com/pages/forms/"

all_rows = []
classes = ["name", "year", "wins", "losses", "ot-losses", "pct", "gf", "ga", "diff"]
page_num = 1

while True:
    page = requests.get(site, params={'page_num': page_num, 'per_page': 100})
    content = BeautifulSoup(page.content, "html.parser")
    teams = content.find_all('tr', class_="team")
    if not teams:
        break
    for team in teams:
        row = [team.find('td', class_=cl).text.strip() for cl in classes]
        all_rows.append(row)
    page_num += 1

headers = [th.text.strip() for th in content.find_all('th')]
```

```python
df = pd.DataFrame(all_rows, columns=headers)
df
```
