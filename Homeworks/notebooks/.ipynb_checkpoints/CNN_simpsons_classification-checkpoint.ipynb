```python
# =========================================================
# 1. Imports & Device
# =========================================================
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, Subset
import pandas as pd
import os
from PIL import Image
import random
import numpy as np
import kornia.augmentation as K
import matplotlib.pyplot as plt
import math

# Device
if torch.cuda.is_available():
    DEVICE = torch.device("cuda")
    print("Using CUDA")
elif torch.backends.mps.is_available():
    DEVICE = torch.device("mps")
    print("Using MPS")
else:
    DEVICE = torch.device("cpu")
    print("Using CPU")

# Seeds
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)
```

```python

# =========================================================
# 2. Paths (AUTO)
# =========================================================
def get_paths():
    if os.path.exists("/content/simpsons"):
        return "/content/simpsons/train", "/content/simpsons/testset"
    elif os.path.exists("../../data/simpsons"):
        return "../../data/simpsons/train", "../../data/simpsons/testset"
    else:
        raise Exception("Dataset not found")


TRAIN_DIR, TEST_DIR = get_paths()
```

```python

# =========================================================
# 3. GPU Augmentations (Kornia)
# =========================================================
mean = torch.tensor([0.485, 0.456, 0.406])
std = torch.tensor([0.229, 0.224, 0.225])

train_aug = torch.nn.Sequential(
    K.RandomResizedCrop((224, 224)),
    K.RandomHorizontalFlip(),
    K.RandomRotation(degrees=15.0),
    K.ColorJitter(0.1, 0.1, 0.1, 0.1),
)

# CPU part (минимум)
base_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])
```

```python

# =========================================================
# 4. Dataset
# =========================================================
base_dataset = datasets.ImageFolder(TRAIN_DIR)

indices = list(range(len(base_dataset)))
random.shuffle(indices)

split = int(0.8 * len(indices))
train_idx = indices[:split]
val_idx = indices[split:]

train_dataset = Subset(
    datasets.ImageFolder(TRAIN_DIR, transform=base_transform),
    train_idx
)

val_dataset = Subset(
    datasets.ImageFolder(TRAIN_DIR, transform=base_transform),
    val_idx
)

pin = True if DEVICE.type == "cuda" else False

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2, pin_memory=pin)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False, num_workers=2, pin_memory=pin)

class_names = base_dataset.classes
```

```python

# =========================================================
# 5. Model
# =========================================================
model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

for param in model.parameters():
    param.requires_grad = False

for param in model.layer4.parameters():
    param.requires_grad = True

model.fc = nn.Linear(model.fc.in_features, len(class_names))
model = model.to(DEVICE)
```

```python

# =========================================================
# 6. Loss & Optimizer
# =========================================================
criterion = nn.CrossEntropyLoss()

optimizer = optim.AdamW([
    {'params': model.layer4.parameters(), 'lr': 1e-4},
    {'params': model.fc.parameters(), 'lr': 1e-3}
])

scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)
```

```python

# =========================================================
# 7. Train / Validate
# =========================================================
def normalize(x):
    return (x - mean[None, :, None, None].to(DEVICE)) / std[None, :, None, None].to(DEVICE)


def train_epoch():
    model.train()
    total_loss, correct, total = 0, 0, 0

    for images, labels in train_loader:
        images = images.to(DEVICE, non_blocking=True)
        labels = labels.to(DEVICE, non_blocking=True)

        images = train_aug(images)
        images = normalize(images)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        total_loss += loss.item() * images.size(0)
        preds = outputs.argmax(1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    return total_loss / total, correct / total


def validate():
    model.eval()
    total_loss, correct, total = 0, 0, 0

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(DEVICE)
            labels = labels.to(DEVICE)

            images = normalize(images)

            outputs = model(images)
            loss = criterion(outputs, labels)

            total_loss += loss.item() * images.size(0)
            preds = outputs.argmax(1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    return total_loss / total, correct / total
```

```python

# =========================================================
# 8. Training
# =========================================================
TRAIN_MODEL = False  # True = обучаем, False = просто используем веса

if TRAIN_MODEL:
    best_loss = float("inf")

    for epoch in range(10):
        train_loss, train_acc = train_epoch()
        val_loss, val_acc = validate()

        scheduler.step()

        print(f"Epoch {epoch + 1}")
        print(f"Train: {train_loss:.4f} | {train_acc:.4f}")
        print(f"Val:   {val_loss:.4f} | {val_acc:.4f}")

        if val_loss < best_loss:
            best_loss = val_loss
            torch.save(model.state_dict(), "CNN_simpsons_classification_model.pth")

# всегда загружаем лучшие веса
model.load_state_dict(torch.load("CNN_simpsons_classification_model.pth"))
```

```python
# =========================================================
# 9. Test Dataset
# =========================================================
class TestDataset(torch.utils.data.Dataset):
    def __init__(self, root, transform):
        self.root = root
        self.transform = transform
        self.images = sorted(os.listdir(root))

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        name = self.images[idx]
        path = os.path.join(self.root, name)

        img = Image.open(path).convert("RGB")
        img = self.transform(img)

        return img, name


test_dataset = TestDataset(TEST_DIR, base_transform)
test_loader = DataLoader(test_dataset, batch_size=32)
```

```python
# =========================================================
# 10. Predict
# =========================================================
model.eval()
results = []

with torch.no_grad():
    for images, names in test_loader:
        images = images.to(DEVICE)
        images = normalize(images)

        outputs = model(images)
        probs = torch.softmax(outputs, 1)
        preds = outputs.argmax(1)

        for i in range(len(preds)):
            results.append({
                "image_id": names[i],
                "prediction": class_names[preds[i]],
                "confidence": probs[i][preds[i]].item()
            })

df = pd.DataFrame(results)
df.to_csv("CNN_simpsons_classification_predictions.csv", index=False)

print(df.head())
```

```python
# =========================================================
# 11. Visualization (Better version)
# =========================================================
def denormalize(img):
    mean_t = mean[None, :, None, None].to(img.device)
    std_t = std[None, :, None, None].to(img.device)
    return img * std_t + mean_t


model.eval()

max_images = 300
cols = 3
rows = math.ceil(max_images / cols)

plt.figure(figsize=(cols * 4, rows * 4))

shown = 0

with torch.no_grad():
    for images, names in test_loader:
        images = images.to(DEVICE)

        outputs = model(normalize(images))
        probs = torch.softmax(outputs, dim=1)
        preds = outputs.argmax(1)

        images = denormalize(images).cpu()

        for i in range(len(images)):
            if shown >= max_images:
                break

            img = images[i].permute(1, 2, 0).numpy()
            img = np.clip(img, 0, 1)

            ax = plt.subplot(rows, cols, shown + 1)
            ax.imshow(img)

            pred_name = class_names[preds[i]]
            confidence = probs[i][preds[i]].item()

            ax.set_title("")  # убираем сверху
            ax.set_xlabel(f"{pred_name}\n{confidence * 100:.1f}%", fontsize=10, labelpad=8)
            ax.xaxis.set_label_position('bottom')
            ax.set_xticks([])
            ax.set_yticks([])

            shown += 1

        # if shown >= max_images:
            break

plt.tight_layout()
plt.show()
```
